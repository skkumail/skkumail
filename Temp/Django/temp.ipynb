```python
user = "tofu7679@g.skku.edu"
password =  "inuv enws vfma ouws"

def find_encoding_info(txt):
    info = email.header.decode_header(txt)
    s, encoding = info[0]
    return s, encoding

imap = imaplib.IMAP4_SSL("imap.gmail.com")
imap.login(user, password)
imap.select("INBOX")
status, messages = imap.search(None, "ALL")
messages = messages[0].split()
recent = messages[-14]
res, msg = imap.fetch(recent, "(RFC822)")
raw = msg[0][1].decode('utf-8')


email_message = email.message_from_string(raw)
print(email_message['From'])
print(email_message['Date'])

message = ''
if email_message.is_multipart():
    for part in email_message.get_payload():
        if part.get_content_type() == 'text/plain':
            bytes = part.get_payload(decode=True)
            encode = part.get_content_charset()
            message += str(bytes, encode)
else:
    if email_message.get_content_type() == 'text/plain':
        bytes = email_message.get_payload(decode=True)
        encode = email_message.get_content_charset()
        message += str(bytes, encode)
print(message)
```
